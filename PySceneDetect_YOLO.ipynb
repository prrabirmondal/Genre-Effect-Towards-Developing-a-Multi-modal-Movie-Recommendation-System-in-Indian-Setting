```python
# This function is for splitting the video into different scenes

def split_video_into_scenes(video_path,output_path, threshold=27.0):
    # Open our video, create a scene manager, and add a detector.
    # print("running\n")
    video = open_video(video_path)
    scene_manager = SceneManager()
    scene_manager.add_detector(
        ContentDetector(threshold=threshold))
    scene_manager.detect_scenes(video, show_progress=True)
    scene_list = scene_manager.get_scene_list()
    save = output_path
    split_video_ffmpeg(video_path, scene_list, show_progress=True,output_file_template=save + '$VIDEO_NAME-Scene-$SCENE_NUMBER.mp4')
```

```python
def model_analysis(frame_folder_path): #frame_folder_path where the frames of a single scene are stored
    
    
#         '''
#             1. select the object having confidence score>=50
#             2. save the frame having heighest number of class and object following the step 1
#             3. save the class wise object frequency of the key frame and save the record in a dictionary
#             4. return dictionary where class name as key and object frequency as value
#         '''

    key_frame = 'none'
    obj_dict = {}
    max_obj_cls = 0
    for key in range(80):
        obj_dict[key]=0
        
        
    for frame in os.listdir(frame_folder_path):        
        # loops for frame
        k_f = frame
        
        frame = os.path.join(frame_folder_path, frame)
        # print(frame)
        results = model5x6(frame)
        
        predict = results.pred
        predict = predict[0]

        obj_cls = []
        
        for obj in range(len(predict)):
            # loop for object
            
            obj_class = int(predict[obj][5].item())            
            obj_percentage = predict[obj][4].item()*100

    # consider objects with more than 50% confidence score
            if obj_percentage > 50:
                obj_cls.append(obj_class)
                
#         print(f'frame = {frame}, total class = {len(set(obj_cls))}, total objects = {len(obj_cls)}')
              
              
        if max_obj_cls < (len(obj_cls)*len(set(obj_cls))):
        # key_frames with heighest confident objects
            key_frame = k_f
            max_obj_cls = (len(obj_cls)*len(set(obj_cls)))
            obj_dict = obj_dict.fromkeys(obj_dict, 0)
        # class_wise object frequencies in key_frame
            for ob_cls in obj_cls:
                obj_dict[ob_cls] += 1
                
        
    if key_frame == 'none':
        obj_dict = obj_dict.fromkeys(obj_dict, 0)
        
    obj_list = [0]*80
    for indx in range(len(obj_list)):
        obj_list[indx] = obj_dict[indx]
        
        
                
    return (key_frame, obj_list)
```

```python
def sum_list(l):
  lp = [0 for i in range(len(l[0]))]
  for i in range(len(l)):
    for j in range(len(l[i])):
      lp[j] += l[i][j]
  return lp
```

```python
# This function is for extracting all the frames of a particular scene

def get_frames(video_path,path):
    # print("running hello\n")
    video_name = os.path.basename(video_path).split('/')[-1]
    len_cal = (len(video_path) - len(video_name) ) + 9
    parrent_path = video_path[0:len_cal] + "_frames"
    os.makedirs(parrent_path)
    parent_save = parrent_path
    parent_dir = os.listdir(path)
    # parent_save = "/content/drive/MyDrive/scene_detect/save_frames"
    movie_names = []

    for child_dir in parent_dir :
        os.makedirs(os.path.join(parent_save,child_dir))       
        count = 1     
        frame_path = parent_save+"/"+child_dir
        #os.mkdir(frame_path)

        video_path = os.path.join(path ,child_dir)
        movie_names.append(child_dir)    
        video_reader = cv2.VideoCapture(video_path)
        #frames_req = 5000           #No. of frame requires from each video 
        video_frames_count = int(video_reader.get(cv2.CAP_PROP_FRAME_COUNT))  #COUNTING NO. OF FRAMES IN A VIDEO
        print(child_dir," ",video_frames_count)


        skip_frames = 1

        for frame_id in range(video_frames_count):
            video_reader.set(cv2.CAP_PROP_POS_FRAMES, frame_id * skip_frames)
            ret, frame = video_reader.read()
            save_path = os.path.join(frame_path,"frame"+ str(frame_id)+".jpg")        
            cv2.imwrite(save_path, frame)

        count += 1        
        # print(count)    
    
    return parrent_path
```

```python
# This function is for extracting the keyframe from a particular scene using YOLO

def get_frames_for_each_scene(video_path,main_folder,object_path):
  
  video_name = os.path.basename(video_path).split('/')[-1]
  len_cal = (len(video_path) - len(video_name) ) + 9
  parrent_path = video_path[0:len_cal] + "_scenes/"
  os.makedirs(parrent_path)
  output_path = parrent_path
  split_video_into_scenes(video_path,output_path)
  frames_path = get_frames(video_path,output_path)
  base = os.path.basename(video_path).split('/')[-1] 
  upper_path_name = main_folder + base[0:9] + "_key_frames"
  os.makedirs(upper_path_name)
  p = os.listdir(frames_path)

  vec = []
  # p = os.listdir("/content/drive/MyDrive/scene_detect/tt3449292_frames")
  for i in p:
     fi = os.path.join(frames_path,i)
     number_frame = len(os.listdir(fi))
     if number_frame > 20:
         get = model_analysis(fi)
         base = os.path.basename(fi).split('/')[-1] 
         path_name = base[0:9] + "_" + base[-7:-4] + "_key_frame"
         save_path = os.path.join(upper_path_name,path_name+".jpg") 
    #  print(get)
         print(get[0]) 
         vec.append(get[1])
         if get[0] != 'none':
             img = cv2.imread(os.path.join(fi,get[0]))
    #  print(get)
    #  print(get[0])      
             cv2.imwrite(save_path, img)
    #  print("Saved: ",save_path) 
   
  g = sum_list(vec)
  obj_path = os.path.join(object_path,base[0:9] + ".obj")
  file_pi = open(obj_path, 'wb') 
  pickle.dump(g, file_pi)
  print("saved file")
  return g
```
